```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import  time
today=time.localtime()

PATH = "https://finance.naver.com"

resp = requests.get(PATH)
soup = BeautifulSoup(resp.text,"lxml")

# . -> 클래스
# # -> id
#  -> tag ( span, li, ul, div )
list = []
link = []
for i in soup.select(".section_strategy li a"):
    list.append(i.text)
    link.append(urljoin(PATH,i["href"]))


df = pd.DataFrame({"타이틀":list,"링크":link})
df.to_excel(f"output/{today.tm_year}_{today.tm_mon}_test.xlsx",index=False)
    
```

```python

result = soup.select('.popular li a')
print(len(result))
print(result)
```
